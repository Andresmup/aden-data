# FAKE DATA CREATOR

```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime 
import os
```

```python
faker = Faker(locale = 'es')
```

## Product

```python
def getProductName():
    product_pool = [
        'TV LG', 'Iphone 7', 'Horno', 'Las aventuras de peter pan', 'Tocadiscos clasico',
        'Tensiometro', 'Pantalon Jean', 'GPU Nvidia', 'Laptop Dell', 'Microondas Samsung',
        'Cafetera Nespresso', 'Aire acondicionado Mitsubishi', 'Bicicleta de montaña',
        'Smartwatch Apple', 'Cámara Canon', 'Tablet Samsung Galaxy', 'Licuadora Oster',
        'Zapatos deportivos Nike', 'Frigorífico Whirlpool', 'Teclado mecánico Logitech',
        'Audífonos Bose', 'Libro de cocina italiana', 'Lavadora Bosch', 'Monitor ASUS 27"',
        'Silla ergonómica', 'Proyector Epson', 'Batidora KitchenAid', 'Altavoz inteligente Alexa',
        'Impresora HP', 'Reloj Casio G-Shock', 'Cortadora de césped Honda', 'Consola Xbox Series X',
        'Cámara de seguridad Ring', 'Guitarra Fender Stratocaster', 'Taladro Bosch', 'Barra de sonido Sony',
        'Patines en línea Rollerblade', 'Despertador digital Philips', 'Router Wi-Fi TP-Link', 'Freidora de aire Philips'
    ]
    random_product = random.randint(0, len(product_pool)-1)
    product_id = product_pool[random_product]
    
    return product_id
```

```python
rows = []
for _ in range(1000):
    name = getProductName()
    price = random.randrange(1, 100)
    category_id = random.randint(1, 10)
    rows.append({
        "name": name,
        "price": price,
        "category_id": category_id
    })

df_product = pd.DataFrame(rows)

df_product
```

## Customer

```python
rows = []
for _ in range(100):
    name = faker.company()
    phone = faker.phone_number()
    address = faker.street_address()
    country = faker.country()
    rows.append({
        "name": name,
        "phone": phone,
        "address": address,
        "country": country
    })

df_customer = pd.DataFrame(rows)

df_customer
```

## Users

```python
rows = []
for _ in range(100):
    name = faker.name()
    mail = faker.email()
    role = faker.job()
    rows.append({
        "name": name,
        "mail": mail,
        "role": role
    })

df_users = pd.DataFrame(rows)

df_users
```

## Sale Order

```python
def getStatus():
    status_pool = [
        'completado', 'pendiente', 'cancelado'
    ]
    random_status = random.randint(0, len(status_pool)-1)
    status = status_pool[random_status]
    
    return status
```

```python
rows = []
for _ in range(100):
    order_date = faker.date_between_dates(date_start=datetime(2022,1,1), date_end=datetime(2024,9,3))
    status = getStatus()
    total_amount = random.randrange(10, 1000)
    customer_id = _+1
    user_id = _+1
    rows.append({
        "order_date": order_date,
        "status": status,
        "total_amount": total_amount,
        "customer_id": customer_id,
        "user_id": user_id
    })

df_sales_order = pd.DataFrame(rows)

df_sales_order
```

## Invoice

```python
def getInvoice():
    invoice_pool = [
        'pagado', 'pendiente', 'expirado'
    ]
    random_invoice = random.randint(0, len(invoice_pool)-1)
    invoice = invoice_pool[random_invoice]
    
    return invoice
```

```python
rows = []
for _ in range(100):
    invoice_date = faker.date_between_dates(date_start=datetime(2022,1,1), date_end=datetime(2024,9,3))
    total_amount = random.randrange(10, 1000)
    due_date = faker.date_between_dates(date_start=datetime(2022,1,1), date_end=datetime(2024,9,3))
    status_invoice = getInvoice()
    order_id = _+1
    product_id = random.randint(0, 999)
    rows.append({
        "invoice_date": invoice_date,
        "total_amount": total_amount,
        "due_date": due_date,
        "status_invoice": status_invoice,
        "order_id": order_id,
        "product_id": product_id
    })

df_invoice = pd.DataFrame(rows)

df_invoice
```

## Payment

```python
def getMethod():
    method_pool = [
        'efectivo', 'tarjeta credito', 'tarjeta debito', 'transferencia bancaria', 'paypal', 'cheque', 'mercado pago'
    ]
    random_method = random.randint(0, len(method_pool)-1)
    method = method_pool[random_method]
    
    return method
```

```python
def getPayment():
    payment_pool = [
        'pagado', 'pendiente', 'fallido'
    ]
    random_payment = random.randint(0, len(payment_pool)-1)
    payment = payment_pool[random_payment]
    
    return payment
```

```python
rows = []
for _ in range(100):
    payment_date = faker.date_between_dates(date_start=datetime(2022,1,1), date_end=datetime(2024,9,3))
    amount = random.randrange(10, 1000)
    method = getMethod()
    status_payment = getPayment()
    payment_installment = random.randint(1, 12)
    invoice_id = random.randint(1, 100)
    rows.append({
        "payment_date": payment_date,
        "amount": amount,
        "method": method,
        "status_payment": status_payment,
        "payment_installment": payment_installment,
        "invoice_id": invoice_id
    })

df_payment = pd.DataFrame(rows)

df_payment
```

## Write data into CSV

```python
dfs = [df_category, df_product, df_customer, df_users, df_sales_order, df_invoice, df_payment]
rutas = ["df_category.csv", "df_product.csv", "df_customer.csv", "df_users.csv", "df_sales_order.csv", "df_invoice.csv","df_payment.csv"]

for d in range(len(rutas)):
    ruta_csv = rutas[d-1]
    df = dfs[d-1]
    if os.path.exists(ruta_csv):
        df.to_csv(ruta_csv, index=False)
        print(f"Archivo CSV existente '{ruta_csv}' sobrescrito correctamente.")
    else:
        df.to_csv(ruta_csv, index=False)
        print(f"Nuevo archivo CSV '{ruta_csv}' creado correctamente.")
```
